```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
r = 5
in_circle = []
in_square = []
iterations = 5000
replication = 5
result = []
cx,cy = 1,2
```

```python
for i in range(replication):
    for i in range(iterations):
        x = np.random.uniform(-r,r)
        y = np.random.uniform(-r,r)
        if math.pow(x,2) + math.pow(y,2) <= math.pow(r,2):
            in_circle.append((x,y))
        else:
            in_square.append((x,y))

    area = math.pow((2 * r),2) * len(in_circle) / (len(in_circle) + len(in_square))
    expected = math.pi * math.pow(r,2)
    result.append(area)

_title = "After {:,} iterations, expected area is {:.2f}\n Prediction Mean:{:.2f} Stdev:{:.3f}".format(iterations,
                                                                                                       expected,
                                                                                                       np.mean(result),
                                                                                                       np.std(result))

plt.figure(figsize=(5, 5))
plt.title(_title)
plt.scatter([x[0]+cx for x in in_circle],[x[1]+cy for x in in_circle],marker=".",alpha=0.5,color='red')
plt.scatter([x[0]+cx for x in in_square],[x[1]+cy for x in in_square],marker=".",alpha=0.5,color='blue')
plt.scatter(cx,cy,marker=".")
plt.grid(True)
plt.show()
```
